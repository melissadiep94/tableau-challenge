```python
data_2020 = pd.read_csv("JC-202009-citibike-tripdata.csv")

data_2020["year"] = "Sept 2020"
#data_2020
```

```python
#Output pandas dataframe into CSV file
output_path = os.path.join('citibike_Sept_2018_to_2020_combined.csv')
print(output_path)
data_combined.to_csv(output_path)
```
